## Exercise 1: Prime Numbers

```python
num=int(input("Enter the Integer"))
if num==0 or num==1:
    print(f"{num} is not a prime number")
else:
    for i in range(2,num):
        if num%i==0:
            print(f"{num}is not a prime number")
            break
        else:
            print(f"{num}is a prime number")
```

## Exercise 2: Product of Random Numbers

```python
import random
rand1=random.randint(1,10)
rand2=random.randint(1,10)
print(f"The product of the {rand1}*{rand2}:")
product=int(input("Enter the Product"))
if product==(rand1*rand2):
    print(f"Your answer is True.The Product is {rand1*rand2}")
else:
    print(f"Your answer is False.The product is {rand1*rand2}")
```

## Exercise 3: Square of even/ odd numbers

```python
print("Square of the even numbers in range of 100 to 200 is:")
for i in range(100,200):
    even= i%2
    if even==0:
        print(f"The square of {i}*{i}={i*i}")
```

## Exercise 4: Hospital Billing System

```python
room=int(input("Enter the number of days stayed"))
lab=int(input("Enter the number of tests has taken"))
med=int(input("Enter the cost of the Medicines"))
room_cost=room*2000
test_cost=lab*300
total_cost=room_cost+test_cost+med+1500
if total_cost > 10000:
    dis=  total_cost*0.1
    print(total_cost-dis)
else:
    print(total_cost)
```
