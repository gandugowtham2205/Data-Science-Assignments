## Descriptive Analytics for Numerical Columns

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dis=pd.read_csv("sales_data_with_discounts.csv")
dis
```

```python
dis_num=dis.select_dtypes(include="number")
dis_num
```

```python
dis_num.mean()
```

```python
dis_num.median()
```

```python
dis_num.mode()
```

```python
np.std(dis_num)
```

### brief interpretation of these statistics.
Most people buy only a small number of things at low prices. Sometimes, people buy expensive items, so the average amount goes up.On average, each sale has about 5 items. Most items are not costly, but a few are very expensive.The discount given is usually around 15%, which means the price is a bit less for most things.After the discount, the final amount changes a lot because some people buy more or costlier things.So, most sales are small and regular. Only a few are very big, which makes the total numbers look very different.

## Data Visualization

```python
for col in dis_num:
    dis[col].hist(bins=20)
    plt.title(f"Histgram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

Distribution: The numbers in our data are spread out. Most of them are close together, but a few are much bigger or much smaller.
Skewness: When many scores are low but some are very high, the chart looks tilted. That means most things are similar but a few are special.
Outliers: Outliers are numbers that are very different from most others. For example, if most prices are around ₹10,000, but one is ₹50,000, that ₹50,000 is an outlier.

```python
dis.boxplot(vert=False,figsize=(10,5),color="green")
```

Sometimes in our data, most numbers are close together, but a few are much bigger or much smaller. These special numbers are called extreme values or outliers.We also see that sometimes most numbers are small, and only a few are big. This is unusual and is called a “skewed” distribution.

```python
dis_cat=dis.select_dtypes(include="object")
dis_cat
```

```python
for col in dis_cat:
    plt.figure(figsize=(10, 4))
    dis[col].value_counts().plot(kind='bar')
    plt.title(f'Frequency Count for {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

Some categories, like brands or models, are seen many times in our data. This means they are very popular many people buy them.Other categories are seen only a few times. They are not very popular or are new.The most common categories are liked by many people.The less common categories may not be interesting to people, or maybe stores need to advertise them more.

### Summary

Some columns have data that is shifted to one side (skewed), which means there are more small values and only a few very large ones.Popular categories, like some brands or models, show up many times, meaning customers like them more.Some categories only appear a few times, which means they are not popular or are new.By looking for outliers and which categories are common, shops and companies can know what people buy most and where they need to improve.Most things in your data are normal, but there are a few big or small numbers and some very popular items. This knowledge will help businesses sell better and find what needs attention.
